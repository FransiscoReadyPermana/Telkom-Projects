# Membuat kolom wa, sms, dan email
nilai-nilai pada kolom tersebut saya dapatkan berdasarkan nilai wildcard.

untuk kolom wa, saya membuat nilai wildcard dengan mencari apakah ada kata what pada service yang ada.
pada kolom sms saya menggunakan nilai wildcard sms, dan pada kolom email saya menggunakan nilai wildcard email

```python
service_columns = [f'Service {x}' for x in range(1, 11)]

df['wa'] = df[service_columns].apply(lambda row: int(any(row.str.contains('what', case=False))), axis=1)
df['sms'] = df[service_columns].apply(lambda row: int(any(row.str.contains('sms', case=False))), axis=1)
df['email'] = df[service_columns].apply(lambda row: int(any(row.str.contains('email', case=False))), axis=1)
df
```

setelah membuat ketiga kolom wa, email, dan sms. selanjutnya, saya menghapus kolom service karena kolom tersebut sudah tidak mengandung informasi apa-apa lagi. kemudian saya membuat kolom total yang nilainya berisi hasil penjumlahan dari kolom sms, wa, dan email

```python
df.drop(columns=service_columns, inplace=True)
df.set_index("Timestamp", inplace=True)
df['total'] = df['sms'] + df['wa'] + df['email'] 
df
```

## Pembuatan Model LSTM
Kemudian saya membuat model lstm buat coba prediksi 4 nilai itu (wa, email, sms, total)

saya Menggunakan 8 sequence length, 1 forecast horizon, 64 jumlah neuron, dan 100 epoch

```python
sequence_length = 8  # Length of the input sequence
forecast_horizon = 1  # Number of steps to forecast ahead


# Prepare data for training
X_train, y_train = [], []
for i in range(len(timeseries) - sequence_length - forecast_horizon + 1):
    X_train.append(timeseries[i:i+sequence_length])
    y_train.append(timeseries[i+sequence_length:i+sequence_length+forecast_horizon])
X_train = np.array(X_train)
y_train = np.array(y_train)

# Build the LSTM model
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(sequence_length, timeseries.columns.shape[0]), return_sequences=False))
model.add(Dense(forecast_horizon * timeseries.columns.shape[0]))  # 4 outputs for 4 value columns

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=16, shuffle=True)

# Make predictions
new_data = timeseries.iloc[-sequence_length:].values
new_data = np.expand_dims(new_data, axis=0)
predictions = np.round(model.predict(new_data))

# Calculate metrics
actual_values = timeseries.iloc[-1].values
mse = mean_squared_error(actual_values, predictions[0])
mape = mean_absolute_percentage_error(actual_values, predictions[0])
r2 = r2_score(actual_values, predictions[0])

print("MSE:", mse)
print("MAPE:", mape)
print("R2:", r2)
```

hasilnya cukup bagus juga ternyata ðŸ‘ðŸ‘
